```python
EGS_total = pd.concat([rdsftnew,mass,prop], axis=1)  #EGS_total la tabla completa, con redshift, masas y propiedades
EGS_total.shape
cols = [
    "z_best",
    "z_best_type",
    "Hmag",
    "EBV_6a_tau_NEB",
    "BMag_6a_deltau",
    "VMag_6a_deltau",
    "logMstar_neb",
    "logMstar_neb_err",
    "SFR_6a_tau_NEB",
    "age_6a_tau_NEB",
    "met_6a_deltau"]

EGS_clean = EGS_total[cols].copy()    #EGS_clean es la tabla con solo las 11 propiedades que nos interesan en principio
EGS_clean.columns = ["z", "z_type", "Hmag", "Color_B-V","M-B","M-V","log_Mass", "err_log_M", "SFR","Age","gasMet"]
#Parametros observados: z, Hmag, ColorBV, M-B, M-V
#Parametros derivados: logMass, SFR, Age, Gas Metalicity
```

```python
#Filtramos solo los valores 1.5<z<2.5 (cosmic noon) que hay dentro de EGS_clean dentro de EGS_cosmic. De 41457 galaxias pasamos a 15150
#Tenemos que filtrar algunos problematicos "no físicos" en Age, Hmag, SFR y Color

EGS_cosmic = EGS_clean[(EGS_clean["z"] > 1.5) & (EGS_clean["z"] < 2.5)].copy()
EGS_cosmic = EGS_cosmic[(EGS_cosmic["Age"] > 0) & (EGS_clean["Age"] < 14)].copy()
EGS_cosmic = EGS_cosmic[(EGS_cosmic["Color_B-V"] > -90)].copy()
EGS_cosmic = EGS_cosmic[(EGS_cosmic["SFR"] > 0)].copy()
EGS_cosmic = EGS_cosmic[(EGS_cosmic["Hmag"] > 0)].copy()

EGS_cosmic.to_csv("candels_EGS_cosmic.csv", index=False)
```

```python
cols_corr = [
    "z",
    "Hmag",
    "Color_B-V",
    "M-B",
    "M-V",
    "log_Mass",
    "SFR",
    "Age",
    "gasMet"]

EGS_corr = EGS_cosmic[cols_corr].copy()
EGS_corr["SFR"] = np.log10(EGS_corr["SFR"])   #Aqui cambiamos SFR por logSFR
EGS_corr.columns = ["z", "Hmag","Color_B-V","M-B","M-V", "log_Mass", "log_SFR", "Age", "gasMet"]
corr_matrix = EGS_corr.corr(method="spearman")

plt.figure(figsize=(6,5))

im = plt.imshow(corr_matrix, vmin=-1, vmax=1)

plt.colorbar(im, label="Coeficiente de correlación")

labels = EGS_corr.columns

plt.xticks(range(len(labels)), labels, rotation=45)
plt.yticks(range(len(labels)), labels)

for i in range(len(labels)):
    for j in range(len(labels)):
        value = corr_matrix.iloc[i, j]
        plt.text(j, i, f"{value:.2f}",
                 ha="center", va="center",
                 fontsize=9,
                 color="black")
        
plt.tight_layout()
plt.savefig("correlation_matrix.pdf")
plt.show()
```

```python

fig, axes = plt.subplots(2, 3, figsize=(12, 6))
plt.subplots_adjust(hspace=0.35, wspace=0.25)
sns.histplot(EGS_cosmic["z"], bins=20,color="grey",edgecolor="black",linewidth=0.8, ax=axes[0,0])
sns.histplot(EGS_cosmic["log_Mass"], bins=30,color="grey",edgecolor="black",linewidth=0.8, ax=axes[0,1])
sns.histplot(EGS_cosmic["Age"], bins=30,color="grey",edgecolor="black",linewidth=0.8, ax=axes[0,2])
sns.histplot(EGS_cosmic["Color_B-V"], bins=23,color="grey",edgecolor="black", ax=axes[1,0])
sns.histplot(log_sfr, bins=30,color="grey",edgecolor="black",linewidth=0.8, ax=axes[1,1])
sns.histplot(EGS_cosmic["Hmag"], bins=25,color="grey",edgecolor="black",linewidth=0.8, ax=axes[1,2])

axes[0,0].set_ylabel("Número de galaxias")
axes[0,1].set_ylabel("")
axes[0,2].set_ylabel("")
axes[1,0].set_ylabel("Número de galaxias")
axes[1,1].set_ylabel("")
axes[1,2].set_ylabel("")

axes[1,1].set_xlabel("log(SFR)")

fig.suptitle("Caracterización de galaxias en el cosmic noon (Campo EGS)")


def add_stats(ax, data):  #Funcion que añade medianas y cuartiles
    median = np.median(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)

    ax.axvline(median, color='red', linewidth=1)
    ax.axvline(q1, color='black', linestyle='--',linewidth=1)
    ax.axvline(q3, color='black', linestyle='--',linewidth=1)

add_stats(axes[0,0], EGS_cosmic["z"])
add_stats(axes[0,1], EGS_cosmic["log_Mass"])
add_stats(axes[0,2], EGS_cosmic["Age"])
add_stats(axes[1,0], EGS_cosmic["Color_B-V"])
add_stats(axes[1,1], log_sfr)
add_stats(axes[1,2], EGS_cosmic["Hmag"])

plt.savefig("panel_histogramas_stats.pdf")

plt.tight_layout()
plt.show()

```
